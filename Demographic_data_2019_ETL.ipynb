```python
df = pd.read_excel(f'{file_path}2019 County Health Rankings Data - v3.xls', header=1, sheet_name=5)
df.head()
```

```python
df_demographics = pd.DataFrame(data={"FIPS":df['FIPS'],
                                     "State":df['State'],
                                     "County":df['County'],
                                     "Population":df['Population'],
                                     "% < 18 Yrs Old":df['% < 18'],
                                     "% 65 Yrs Old and over": df['% 65 and over'],
                                     "Income": df['Household Income'],
                                     "African American (Count)": df['# African American'],
                                     "African American (%)": df['% African American'],
                                     "American Indian/Alaskan Native (Count)": df['# American Indian/Alaskan Native'],
                                     "American Indian/Alaskan Native(%)": df['% American Indian/Alaskan Native'],
                                     "Asian (Count)": df['# Asian'],
                                     "Asian (%)": df['% Asian'],
                                     "Native Hawaiian/Other Pacific Islander (Count)":df['# Native Hawaiian/Other Pacific Islander'], 
                                     "Native Hawaiian/Other Pacific Islander (%)":df['% Native Hawaiian/Other Pacific Islander'], 
                                     "Hispanic (Count)": df['# Hispanic'], 
                                     "Hispanic (%)": df['% Hispanic'], 
                                     "Non-Hispanic White (Count)":df['# Non-Hispanic White'], 
                                     "Non-Hispanic White (%)": df['% Non-Hispanic White'], 
                                     "Female (%)": df['% Female'], 
                                     "Rural (Count)": df['# Rural'], 
                                     "Rural (%)": df['% Rural'],
                                     "Life Expectancy": df['Life Expectancy'], 
                                     "Life Expectancy (Black)": df['Life Expectancy (Black)'], 
                                     "Life Expectancy (Hispanic)": df['Life Expectancy (Hispanic)'], 
                                     "Life Expectancy (White)": df['Life Expectancy (White)'], 
                                     "Deaths (Count)": df['# Deaths']
                                    })
print(df_demographics.shape)
df_demographics.head(20)
```
