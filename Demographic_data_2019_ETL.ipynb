```python
import os
import pandas as pd
import numpy as np
```

```python
file_path = "Data/"
```

```python
df = pd.read_excel(f'{file_path}2019 County Health Rankings Data - v3.xls', header=1, sheet_name=5)
df.head()
```

```python
for col in df.columns:
    print(col)
```

```python
df_demographics = pd.DataFrame(data={"FIPS":df['FIPS'],
                                     "State":df['State'],
                                     "County":df['County'],
                                     "Population":df['Population'],
                                     "%<18_Yrs_Old":df['% < 18'].map("{:.2f}".format),
                                     "%65_Yrs_Old_and_over": df['% 65 and over'].map("{:.2f}".format),
                                     "Income($)": df['Household Income'].map("${:.2f}".format),
                                     "African_American(Count)": df['# African American'],
                                     "African_American(%)": df['% African American'].map("{:.2f}".format),
                                     "American_Indian/Alaskan_Native(Count)": df['# American Indian/Alaskan Native'],
                                     "American_Indian/Alaskan_Native(%)": df['% American Indian/Alaskan Native'].map("{:.2f}".format),
                                     "Asian(Count)": df['# Asian'],
                                     "Asian(%)": df['% Asian'].map("{:.2f}".format),
                                     "Native_Hawaiian/Other_Pacific_Islander(Count)":df['# Native Hawaiian/Other Pacific Islander'], 
                                     "Native_Hawaiian/Other_Pacific_Islander(%)": df['% Native Hawaiian/Other Pacific Islander'].map("{:.2f}".format), 
                                     "Hispanic(Count)": df['# Hispanic'], 
                                     "Hispanic(%)": df['% Hispanic'].map("{:.2f}".format), 
                                     "Non-Hispanic_White_(Count)":df['# Non-Hispanic White'], 
                                     "Non-Hispanic_White(%)": df['% Non-Hispanic White'].map("{:.2f}".format), 
                                     "Female(%)": df['% Female'].map("{:.2f}".format), 
                                     "Rural(Count)": df['# Rural'], 
                                     "Rural(%)": df['% Rural'].map("{:.2f}".format),
                                     "Life_Expectancy": df['Life Expectancy'].map("{:.2f}".format),
                                     "Deaths(Count)": df['# Deaths']
                                    })
print(df_demographics.shape)
df_demographics.head(15)
```

```python
df_demographics.dtypes
```

```python
type(df_demographics['FIPS'])
```

```python
lyme_df = pd.read_csv(f'{file_path}LymeCaseCountsbyCounty_FIPS.csv', encoding='latin-1')
lyme_df.head(20)
```

```python
df_lyme_2019 = lyme_df.drop(columns=['STCODE', 'CTYCODE', 'Unique Identifier','Ctyname', 'Cases2000', 'Cases2001',
       'Cases2002', 'Cases2003', 'Cases2004', 'Cases2005', 'Cases2006',
       'Cases2007', 'Cases2008', 'Cases2009', 'Cases2010', 'Cases2011',
       'Cases2012', 'Cases2013', 'Cases2014', 'Cases2015', 'Cases2016','Cases2017','Cases2018'])
df_lyme_2019 = df_lyme_2019.rename(columns={'Ctyname_Updated':'County','Stname':'State','Cases2019':'Lyme_Disease_Incidence_Reported'})
print(df_lyme_2019.shape)
df_lyme_2019.head()
```

```python
df_lyme_2019 = df_lyme_2019.sort_values('FIPS', ascending=False)
print(df_lyme_2019.shape)
df_lyme_2019.head(10)
```

```python
df_lyme_2019.dtypes
```

```python
df_lyme_2019.to_csv('Data/Lyme2019DF.csv', index=False)
```

```python
df_merged = df_lyme_2019.merge(df_demographics, how ='right')


print(df_merged.shape)
df_merged.head()
```

```python
df_merged.dtypes
```

```python
# Check for null values
for col in df_merged:
    print(f'{col} has {df_merged[col].isnull().sum()} nulls')
```

```python
df_merged.loc[df_merged['FIPS'] == 99940]
```

```python
# Drop rows with NaNs
df_merged = df_merged.dropna()
```

```python
print(f'Are there any null values? {df_merged.isnull().values.any()}')
print(df_merged.shape)
df_merged.head(20)
```

```python
# Check for the incidents of lyme disease reported by county
cases_df = df_merged.groupby(['State','County'])
cases_df.sum().head(15)
```

```python
# Export Merged DF as csv
df_merged.to_csv('Data/Merged_Demographic_Data.csv', index=False)
```

### Join methon 'right' was decided to include a majority of the county demographic info and ommit the '999' county codes rows from the lyme incidence file. This results in less NaNs and more overall data for analysis. However, the tradeoff being there will be less Lyme disease incidence repoted data
