```python
import os
import pandas as pd
import numpy as np
```

```python
file_path = "Data/"
```

```python
df = pd.read_excel(f'{file_path}2019 County Health Rankings Data - v3.xls', header=1, sheet_name=5)
df.head()
```

```python
for col in df.columns:
    print(col)
```

```python
df_demographics = pd.DataFrame(data={"FIPS":df['FIPS'],
                                     "State":df['State'],
                                     "County":df['County'],
                                     "Population":df['Population'],
                                     "% < 18 Yrs Old":df['% < 18'],
                                     "% 65 Yrs Old and over": df['% 65 and over'],
                                     "Income ($)": df['Household Income'],
                                     "African American (Count)": df['# African American'],
                                     "African American (%)": df['% African American'],
                                     "American Indian/Alaskan Native (Count)": df['# American Indian/Alaskan Native'],
                                     "American Indian/Alaskan Native(%)": df['% American Indian/Alaskan Native'],
                                     "Asian (Count)": df['# Asian'],
                                     "Asian (%)": df['% Asian'],
                                     "Native Hawaiian/Other Pacific Islander (Count)":df['# Native Hawaiian/Other Pacific Islander'], 
                                     "Native Hawaiian/Other Pacific Islander (%)":df['% Native Hawaiian/Other Pacific Islander'], 
                                     "Hispanic (Count)": df['# Hispanic'], 
                                     "Hispanic (%)": df['% Hispanic'], 
                                     "Non-Hispanic White (Count)":df['# Non-Hispanic White'], 
                                     "Non-Hispanic White (%)": df['% Non-Hispanic White'], 
                                     "Female (%)": df['% Female'], 
                                     "Rural (Count)": df['# Rural'], 
                                     "Rural (%)": df['% Rural'],
                                     "Life Expectancy": df['Life Expectancy'],
                                     "Deaths (Count)": df['# Deaths']
                                    })
print(df_demographics.shape)
df_demographics.head(20)
```

```python
df_demographics.dtypes
```

```python
type(df_demographics['FIPS'])
```

```python
lyme_df = pd.read_csv(f'{file_path}LymeCaseCountsbyCounty_FIPS.csv', encoding='latin-1')
lyme_df.head(20)
```

```python
df_lyme_2019 = lyme_df.drop(columns=['STCODE', 'CTYCODE', 'Ctyname', 'Cases2000', 'Cases2001',
       'Cases2002', 'Cases2003', 'Cases2004', 'Cases2005', 'Cases2006',
       'Cases2007', 'Cases2008', 'Cases2009', 'Cases2010', 'Cases2011',
       'Cases2012', 'Cases2013', 'Cases2014', 'Cases2015', 'Cases2016','Cases2017','Cases2018'])
df_lyme_2019 = df_lyme_2019.rename(columns={'Ctyname_Updated':'County','Stname':'State'})
print(df_lyme_2019.shape)
df_lyme_2019.head()
```

```python
df_lyme_2019 = df_lyme_2019.sort_values('FIPS')
df_lyme_2019.head(10)
```

```python
df_lyme_2019.dtypes
```

```python
df_merged = df_lyme_2019.merge(df_demographics, how ='left')


print(df_merged.shape)
df_merged.head(20)
```

```python
# Check for null values
for col in df_merged:
    print(f'{col} has {df_merged[col].isnull().sum()} nulls')
```

```python
# Drop rows with NaNs
df_merged = df_merged.dropna()
```

```python
print(f'Are there any null values? {df_merged.isnull().values.any()}')
print(df_merged.shape)
df_merged.head(20)
```

```python
# Export Merged DF as csv
df_merged.to_csv('Data/Merged_Demographic_Data.csv')
```
